```python
import math
import matplotlib.pyplot as plt
from tabulate import tabulate 
g=9.8
m=68.1
c=12.5
t=0
t1=2
v=0
ev=0
ta= []
v1=[]
vel2= []
t2=[]
Ev,Er,Ea,tabla=[],[],[],[]
tol=(0.5*math.pow(10,(-3)))
        
for i in range (60):
  vi=(v+(g-(c/m*v))*(t1-t))
  ta.append(t) 
  v1.append(vi)
  v=vi
  t=t1
  t1=t1+2  
    
for i in range (120):   
  if(i%2==0):   
   vta=(((g*m)/c)*(1-math.exp(-(c/m)*i))) 
   vel2.append(vta)
   t2.append(i)


for t in range (60):
    ev=math.fabs(vel2[t]-v1[t])
    Ev.append(ev)
    
for t in range (60): 
   if(t==0):
    Er.append(100)
   if (t>0): 
    er=(Ev[t]/vel2[t])*100
    Er.append(er)
    
for t in range (59):
 if (t==0):
  Ea.append(100)
  print ("Ts   | velocidad Aprox  | Error relativo % | Error Aproximado %")
 Ea.append(((v1[t+1]-v1[t])/(v1[t+1]))*100) 
 print("%s    |   %f     |     %f    |     %f"%(ta[t+1],v1[t],Er[t+1],Ea[t+1]))
  



plt.title("errores")
plt.plot(ta, Er,'*')
plt.plot(ta, Ea)
plt.plot (ta, Er,label='Error relativo porcentual verdadero')
plt.plot (ta, Ea,label='Error porcentual Aproximado')
plt.grid(True)
plt.axis([0, 28, 0, 40])
plt.legend()
plt.show()
```
