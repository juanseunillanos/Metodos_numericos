```python
#9.3
import math
Mta=[[1, 6],[3, 10],[7, 4]]
Mtb=[[1, 3],[0.5, 2]]
Mtc=[[2, -2],[-3, 1]]

def multiplicacion(m1,m2):
    if ((len(m1[0]))!=(len(m2))):
        print("no se puede realizar la operacion entre matrices ya que no cumple el requisito de multiplicacion es cual dice que el numereo de columnas de la M1 tiene que ser iual al de filas de m2")
    if(len(m1[0])==len(m2)):   
        a=[]
        for i in range(len(m1)):
            a.append([0]*len(m2[0]))
            for c in range(len(m2[0])):
                a[i][c]=0
            
        for x in range(len(m1)):
            for j in range(len(m1[0])):
                for k in range (len(m2[0])):
                    a[x][k]=a[x][k]+(m1[x][j]*m2[j][k])
        print ("el resultado de la multiplicacion es : %s"%(a))
        
print("Mta x Mtb")
multiplicacion(Mta,Mtb)
print ("Mtb x Mta")
multiplicacion(Mtb,Mta)
print ("Mta x Mtc")
multiplicacion(Mta,Mtc)
print("Mtc x Mta")
multiplicacion(Mtc,Mta)
print ("Mtc x Mtb")
multiplicacion(Mtc,Mtb)
print ("Mtb x Mtc")
multiplicacion(Mtb,Mtc)
```
